## <b>Data Cleaning and Preprocessing</b>

#### Addressing missing or null values

```python
df.fillna(df.mean(), inplace=True)
```

#### Transforming categorical data into numerical format using suitable methods.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for column in df.select_dtypes(include='object'):
    df[column] = label_encoder.fit_transform(df[column])

df.head()
```

```python
data =df
```

## <b> Statistical Analysis with NumPy and SciPy</b>

```python
import numpy as np
from scipy import stats

stats_results = {}

for column in data.columns:
    column_data = data[column]
    
    mean = np.mean(column_data)
    median = np.median(column_data)
    mode = stats.mode(column_data)[0]
    
    data_range = np.ptp(column_data) 
    variance = np.var(column_data)
    std_dev = np.std(column_data)
    
   
    skewness = stats.skew(column_data)
    kurtosis = stats.kurtosis(column_data)
    
    # Storing results in the dictionary
    stats_results[column] = {
        'Mean': mean,
        'Median': median,
        'Mode': mode,
        'Range': data_range,
        'Variance': variance,
        'Standard Deviation': std_dev,
        'Skewness': skewness,
        'Kurtosis': kurtosis
    }

# Converting the results dictionary to a DataFrame for better visualization

stats_df = pd.DataFrame(stats_results).T
stats_df.head()

```

```python
# Performing hypothesis testing for statistical significance between groups
# Analysis of variance (ANOVA) to test for differences in means between different price ranges for continuous variables

grouped_data = data.groupby('price_range')

# Dictionary to store the results of the ANOVA test
anova_results = {}

# Iterate through each column (excluding the target 'price_range')
for column in data.columns:
    if column != 'price_range':
        # Collect data for each group
        groups = [group[column].values for name, group in grouped_data]
        
        # Perform Analysis of Variance Test
        f_stat, p_val = stats.f_oneway(*groups)
        
        # Store results
        anova_results[column] = {
            'F-Statistic': f_stat,
            'P-Value': p_val
        }

# Convert the results dictionary to a DataFrame for better visualization
anova_df = pd.DataFrame(anova_results).T
anova_df.head()

```

### Key Observations from ANOVA Results:
Battery Power has a significant p-value (close to 0), indicating that there are statistically significant differences in mean battery power across different price ranges.<br>
Other features like blue, clock_speed, dual_sim, and fc do not show statistically significant differences between price ranges based on their p-values being higher than typical significance levels (e.g., 0.05).

Battery Power is the only feature among those listed with a statistically significant effect on the price_range (low P-Value and high F-Statistic).<br>
Other features like blue (Bluetooth support), clock_speed (processor speed), dual_sim (dual SIM capability), and fc (front camera resolution) do not show significant differences across price_range categories, indicating they are less important or have no impact on the pricing of the phones.

```python
from scipy import stats

spearman_correlation_results = {}

# Iterate through each feature in the dataset (excluding the target variable 'price_range')
for feature in df.columns:
    if feature != 'price_range':
        # Calculate Spearman correlation coefficient and p-value
        correlation, p_value = stats.spearmanr(df[feature], df['price_range'])
        
    
        spearman_correlation_results[feature] = {
            'Spearman Correlation': correlation,
            'P-Value': p_value
        }

# Convert the results dictionary to a DataFrame for better visualization
spearman_corr_df = pd.DataFrame(spearman_correlation_results).T

spearman_corr_df.head()
```

### Observation
Battery Power shows a moderate positive correlation with price_range, suggesting that higher battery power is associated with higher price ranges. <br>
This relationship is statistically significant with a very low p-value.
Other features, such as blue, clock_speed, dual_sim, and fc, have weak or negligible correlations with price_range, and their p-values are not significant, indicating a lack of strong association.

#### Key Determinants in Mobile Price Classification

Both the ANOVA and correlation analyses indicate that RAM is the strongest predictor of mobile price range. Conclusion: <br>

RAM is a crucial determinant of mobile price classification, likely because higher RAM supports better performance, multitasking, and overall device efficiency, which are important factors for higher-priced devices.

##### Battery Power:

The scatter and box plots further confirmed that higher battery power tends to be associated with higher price ranges. The heatmap showed a moderate positive correlation with price_range.<br>
Conclusion: <br>

Battery power is a key determinant, as longer battery life is often a selling point for more expensive models, catering to consumers who need more durable devices.

##### Screen Resolution (px_height and px_width):

Insights: The scatter plots and box plots showed variability in screen resolution across price ranges, but the trends were less distinct compared to RAM and battery power. The correlation heatmap indicated moderate correlations between pixel height and width but weaker correlations with price_range.<br>
Conclusion: <br>
While screen resolution is important for user experience, its impact on price classification is less pronounced than RAM and battery power.

#### <b>Overall Conclusion</b>
The analysis identifies RAM and Battery Power as the most influential determinants in mobile price classification, with other features such as screen resolution having secondary importance. Features like Bluetooth support, dual SIM capability, processor clock speed, and front camera resolution were found to have minimal impact on determining the mobile price range.
